# Question 3 - Should I wait to invest or should I invest now and where?
### (You could recommend more than one place)

I would not recommend investing in the current housing market -- the price has essentially plateaued between 2023 and 2024, as has the home inventory. There's a decent chance that higher interest rates cool the market. 

But in the interest of answering the question with a visual, I will assume that any current local trends in price will continue (which often happens in the housing market!).

### Creating a Price Metric
To understand how markets are moving in the past 6 months, we create a "6_month_change" metric that is the difference between the mean home price in a state in August 2023 vs. February 2024 *as a proportion of the price in August 2023*. Presumably, markets with increasing prices are hot markets, and good to invest in. 

### Standardizing the Inventory 
To understand each how each state's current inventory compares to other states, it doesn't really make sense to compare raw numbers (New York could have a lot more supply than Montana, but still have a shortage relative to homebuyers). While we *could* create a metric of inventory/population, that wouldn't necessarily capture the number of prospective homebuyers, or the historical state of the market. 

Instead, we opted to standardize each state's Feb 29, 2024 inventory over the time in the dataset, in a z-score (Feb 29 inventory - state mean)/(state sd). We will use this as a metric of which states relatively have little supply based on their historical level.

Now we can plot everything.

```python
#mths = pd.date_range(start = '2018-01-01', end = '2024-01-01', freq='MS')

mean_reg_sale_price = mean_sale_price.groupby(["RegionName"]).mean(numeric_only=True)
mean_reg_sale_price_6mo = mean_reg_sale_price.loc[:,["SizeRank","2023-08-31","2024-02-29"]]
mean_reg_sale_price_6mo["6_month_change"] = 100*(mean_reg_sale_price_6mo["2024-02-29"] - mean_reg_sale_price_6mo["2023-08-31"])/mean_reg_sale_price_6mo["2023-08-31"]
mean_reg_sale_price_6mo["price growing?"] = mean_reg_sale_price_6mo["6_month_change"] > 0

reg_inventory = for_sale_listings.drop(["RegionID","SizeRank"], axis=1).groupby(["RegionName"]).sum(numeric_only=True)
reg_inventory_avg = reg_inventory.mean(axis=1)
reg_inventory_sd = reg_inventory.std(axis=1)

reg_inventory = reg_inventory[["2024-02-29"]]
reg_inventory["avg"] = reg_inventory_avg
reg_inventory["sd"] = reg_inventory_sd
reg_inventory["standardized_inventory"] = (reg_inventory["2024-02-29"] - reg_inventory["avg"])/reg_inventory["sd"]

reg_jdf = mean_reg_sale_price_6mo.merge(reg_inventory, on= "RegionName")

```

```python
reg_jdf.loc[["Paris, TN"]]
```

```python
#mths = pd.date_range(start = '2018-01-01', end = '2024-01-01', freq='MS')

mean_state_sale_price = mean_sale_price.groupby(["StateName"]).mean(numeric_only=True)
mean_state_sale_price_6mo = mean_state_sale_price.loc[:,["SizeRank","2023-08-31","2024-02-29"]]
mean_state_sale_price_6mo["6_month_change"] = 100*(mean_state_sale_price_6mo["2024-02-29"] - mean_state_sale_price_6mo["2023-08-31"])/mean_state_sale_price_6mo["2023-08-31"]
mean_state_sale_price_6mo["price growing?"] = mean_state_sale_price_6mo["6_month_change"] > 0

state_inventory = for_sale_listings.drop(["RegionID","SizeRank"], axis=1).groupby(["StateName"]).sum(numeric_only=True)
state_inventory_avg = state_inventory.mean(axis=1)
state_inventory_sd = state_inventory.std(axis=1)

state_inventory = state_inventory[["2024-02-29"]]
state_inventory["avg"] = state_inventory_avg
state_inventory["sd"] = state_inventory_sd
state_inventory["standardized_inventory"] = (state_inventory["2024-02-29"] - state_inventory["avg"])/state_inventory["sd"]

jdf = mean_state_sale_price_6mo.merge(state_inventory, on= "StateName")
```

```python
#fig,ax = plt.subplots()
p = (
so.Plot(data=jdf,x="6_month_change",y="standardized_inventory",text="StateName", color = "price growing?")
    .add(so.Dot())
    .add(so.Text(halign='left'))
    .label(title = "State Price Trend by Inventory Level -- Figure 7")
    .scale(x=so.Continuous().label(like="{x:,.1f}%"))
)

p
```

```python
#fig,ax = plt.subplots()
p = (
so.Plot(data=reg_jdf,x="6_month_change",y="standardized_inventory",text="RegionName", color = "price growing?")
    .add(so.Dot(alpha=.75))
    .add(so.Text(halign='left', alpha = .5))
    .label(title = "Region Price Trend by Inventory Level -- Figure 6")
    .scale(x=so.Continuous().label(like="{x:,.1f}%"))
    .limit(x = (-60,100))
)

p
```

Our plot suggests two possible courses of investment action. 

#### If we are looking for a regional market that is experiencing an extreme price boom, irrespective of how much inventory it has, we should invest in Paris, TN which has seen price increases of ~65% in 6 months. 
6 million dollars would get you ~17 houses in this market, which would be a substantial (~10%) chunk of the average housing stock.

We then need to calculate how much the total costs will be after mortgages. This will be done using the terms of the default mortgage described above.

```python
def total_cost(purchase_price, annual_interest_rate, loan_term_years):
    down_payment = purchase_price * 0.2
    loan_amount = purchase_price - down_payment

    monthly_interest_rate = annual_interest_rate / 100 / 12
    loan_term_months = loan_term_years * 12
    monthly_payment = (loan_amount * monthly_interest_rate) / (1 - (1 + monthly_interest_rate) ** -loan_term_months)

    closing_payment = down_payment
    total_monthly_payment = monthly_payment
    total_yearly_payment = 12 * total_monthly_payment

    return down_payment + loan_term_years * total_yearly_payment
```

Below is a calculation we play around with to idenitfy a number we feel comfortable with. 

```python
def calculate_and_print_costs(purchase_price, annual_interest_rate, loan_term_years, total_money_saved):
    down_payment = purchase_price * 0.2
    loan_amount = purchase_price - down_payment
    closting_costs = 0.05 * purchase_price
    property_tax_rate = (0.0161+0.0161+0.008)/3
    property_tax = property_tax_rate * purchase_price
    home_insurance = 0.005 * purchase_price

    monthly_interest_rate = annual_interest_rate / 100 / 12
    loan_term_months = loan_term_years * 12
    monthly_payment = (loan_amount * monthly_interest_rate) / (1 - (1 + monthly_interest_rate) ** -loan_term_months)

    closing_payment = down_payment + closting_costs
    total_monthly_payment = monthly_payment + property_tax/12 + home_insurance/12
    total_yearly_payment = 12 * total_monthly_payment

    remaining_money_after_closing = total_money_saved - closing_payment
    number_of_payments = remaining_money_after_closing // total_monthly_payment

    remaining_loan_balance = loan_amount - (number_of_payments * monthly_payment)

    print("You will have to pay $", "{:,}".format(closing_payment), " at closing")
    print("Monthly mortgage payment with taxes and insurance included: $", "{:,}".format(round(total_monthly_payment, 2)))
    print("You will have to pay $", "{:,}".format(round(total_yearly_payment, 2)), " per year")
    print("In your first year, you will pay $", "{:,}".format(closing_payment + total_yearly_payment), " in total")
    print("With your savings, you can make ", number_of_payments, " payments")
    print("After one year, you will still owe $", "{:,}".format(round(remaining_loan_balance, 2)))

# Call the function
calculate_and_print_costs(15000000, 7, 15, 6000000)
```
